```python
import numpy as np
from scipy.stats import wasserstein_distance

# -----------------------------------
# Example Real Images
# -----------------------------------

real_images = np.random.rand(
    100,
    28,
    28
)

# -----------------------------------
# Example Generated Images
# Replace with GAN generated images
# -----------------------------------

generated_images = np.random.rand(
    100,
    28,
    28
)

# -----------------------------------
# Flatten Images
# -----------------------------------

real_flat = real_images.reshape(100, -1)

generated_flat = generated_images.reshape(100, -1)

# -----------------------------------
# Compute Wasserstein Distance
# -----------------------------------

distances = []

for i in range(real_flat.shape[1]):

    wd = wasserstein_distance(
        real_flat[:, i],
        generated_flat[:, i]
    )

    distances.append(wd)

# Average Distance
wasserstein_score = np.mean(distances)

print("Wasserstein Distance:",
      wasserstein_score)
```
